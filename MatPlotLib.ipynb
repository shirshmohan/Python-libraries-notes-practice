## Matplotlib example workflow

```python

%matplotlib inline
import matplotlib.pyplot as plt
#1 Prepare data
x  = [1,2,3,4]
y = [11,22,33,44]
#2 Setup plot
fig,ax = plt.subplots(figsize=(10,10)) #(width,height)
#3 Plot data
ax.plot(x,y)
#4 Customize plots
ax.set(title = "Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")
#5 Save and show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

## Which one should you use?(pyplot vs matplotlib)
When plotting something advanced use the OO method ahead

To plotsomething quickly it isokay to use pyplot method

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
fig,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
            ax=ax);
ax.set_xlim([45,100])


```

```python
over_50
```

```python
## OO method from scratch
fig,ax = plt.subplots(figsize=(10,6))

#Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter");
#Ucstomuze the plot
ax.set(title = "Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");
#Add a legend
ax.legend(*scatter.legend_elements(),title="Target");

#Add a horizontal line 
ax.axhline(over_50["chol"].mean(),linestyle='--');
```

```python
over_50.head()
```

```python
#subplotof chol,age,thalach
fig,(ax0,ax1) = plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True);
#Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]);
scatter1 = ax1.scatter(x=over_50["age"],
                      y = over_50["thalach"],
                      c=over_50["target"]);
#customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");
ax1.set(title="Heart Disease and Thalach Levels",
       xlabel="Age",
       ylabel="Max Heart Rate");
#Adding legend
ax0.legend(*scatter.legend_elements(),title="Target");
ax1.legend(*scatter.legend_elements(),title="Target");
#Add a mealine
ax0.axhline(y=over_50["chol"].mean());
ax1.axhline(y=over_50["thalach"].mean());

#Add a titletothe figure 
fig.suptitle("Heart Disease Analysis ", fontsize = 16, fontweight="bold");
```
